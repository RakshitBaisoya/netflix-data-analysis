```python
# importing liberaries
import pandas as pd
import matplotlib.pyplot as plt

#load the data
df=pd.read_csv('netflix_titles.csv')
#print(df.head())
# cleaning data
type=df.isnull().sum()
#print(type)
df=df.dropna(subset=['director','cast','country','date_added','rating','duration'])
type_counts= df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index,type_counts.values,color=['red','pink'])
plt.xlabel('type')
plt.ylabel('counting')
plt.title('number of movies vs tv shows')
plt.savefig('movie_vs_tvshows.png')

plt.show()

rating_counts=df['rating'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(rating_counts,labels=rating_counts.index,autopct='%1.1f%%')
plt.title('percentage of rating')
plt.savefig('rating.png')
plt.show()

movie_df=df[df['type']=='Movie'].copy()
movie_df['duration.int']=movie_df['duration'].str.replace('min','').astype(int)

plt.hist(movie_df['duration.int'],bins=30,color='pink',edgecolor='black')
plt.xlabel('time duration')
plt.ylabel('no_of_movies')
plt.title('duratin of movies')
plt.grid(linestyle=':')
plt.savefig('timing of movies')

plt.show()

release_counts = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.scatter(release_counts.index, release_counts.values)
plt.grid(linestyle=':')
plt.title('Number of Movies Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('Number of Movies Released Each Year')

plt.show()
```
